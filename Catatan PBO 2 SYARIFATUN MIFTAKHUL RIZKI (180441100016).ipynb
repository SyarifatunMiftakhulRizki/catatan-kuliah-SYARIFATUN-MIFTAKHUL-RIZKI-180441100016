# Encapsulation 

Encapsulation adalah sebuah sebuah metode untuk mengatur component class dengan cara menyembunyikan alur kerja dari class tersebut.

Component class yang dimaksud adalah property dan method.

kita bisa membuat pembatasan akses kepada property dan method, sehingga hanya property dan method tertentu saja yang bisa diakses dari luar class.

Dua hal yang mendasar dalam enkapsulasi yakni :

	Information hiding : Sebelumnya kita dapat mengakses anggota class baik berupa atribut maupun method secara langsung dengan menggunakan objek yang telah kita buat. Hal ini dikarenakan akses kontrol yang diberikan kepada atribut maupun method yang ada di dalam class tersebut adalah public. Kita dapat menyembunyikan informasi dari suatu class sehingga anggota class tersebut tidak dapat diakses dari luar, caranya adalah hanya dengan memberikan akses kontrol private ketika mendeklarasikan atribut atau method. Proses ini disebut dengan information hiding.

	Interface to access data : caranya adalah dengan membuat suatu interface berupa method untuk menginisialisasi atau merubah nilai dari suatu atribut tersebut.

Konsep enkapsulasi sangat penting dilakukan untuk menjaga kebutuhan program agar dapat diakses sewaktu-waktu, sekaligus menjaga program tersebut.




Pendefinisian Atribute

__nama : rifa      #pendeklarasian variable private
Dengan mendeklarasikan suatu variable yang didahului dengan garis bawah dua kali, maka hal tersebut berarti bahwa variable tersebut merupakan variable private.

Nama : rifa         #pendeklarasian variable public
Pendeklarasian variable tanpa didahului garis bawah dua kali, maka variable tersebut adalah variable public.
